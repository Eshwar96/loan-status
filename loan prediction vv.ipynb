```python
def replacer(df):
    import pandas as pd
    Q = pd.DataFrame(df.isna().sum(),columns=["vaibhav"])
    for i in Q[Q.vaibhav > 0].index:
        if(df[i].dtypes == "object"):
            x = df[i].mode()[0]
            df[i] = df[i].fillna(x)
        else:
            x = df[i].mean()
            df[i] = df[i].fillna(x)
```
